# **Ejercicio Resuelto N°4**



A la mamá de Juan le preguntan su edad, y contesta: tengo 3 hijos, pregúntele a Juan su edad. Alberto tiene 2/3 de la edad de Juan, Ana tiene 4/3 de la edad de Juan y mi edad es la suma de las tres. Hacer un algoritmo que muestre la edad de los cuatro.

````python
class persona:

  def __init__(self, nombre, edad =0.0):

    self.nombre = nombre
    self.edad = float(edad)


  def mostrar(self):

    print(f"{self.nombre}:{int(self.edad)} años")


class Familia:

  def __init__(self, edad_Juan):

    self.Juan = persona("Juan", edad_Juan)
    self.Alberto = persona("Alberto", (2/3) * self.Juan.edad)
    self.Ana = persona("Ana",(4/3) * self.Juan.edad)
    self.Mama = persona("Mamá",self.Juan.edad + self.Alberto.edad+self.Ana.edad)


  def mostrar_edades(self):

      self.Juan.mostrar()
      self.Alberto.mostrar()
      self.Ana.mostrar()
      self.Mama.mostrar()



edad_Juan = float(input("Ingrese la edad de Juan: "))


familia_instancia = Familia(edad_Juan)

familia_instancia.mostrar_edades()
````

# **Ejercicio Propuesto N° 12**

Un empleado trabaja 48 horas en la semana a razón de $5.000 hora. El porcentaje de
retención en la fuente es del 12,5% del salario bruto. Se desea saber cuál es el salario bruto,
la retención en la fuente y el salario neto del trabajador.

````python
class Empleado:
    def __init__(self, horas, valor_hora, porcentaje_retencion):
        self.horas = horas
        self.valor_hora = valor_hora
        self.porcentaje_retencion = porcentaje_retencion / 100

    def calcular_salario_bruto(self):
        return self.horas * self.valor_hora

    def calcular_retencion(self):
        return self.calcular_salario_bruto() * self.porcentaje_retencion

    def calcular_salario_neto(self):
        return self.calcular_salario_bruto() - self.calcular_retencion()

    def mostrar_resultados(self):
        print("Salario bruto: $", self.calcular_salario_bruto())
        print("Retención en la fuente: $", self.calcular_retencion())
        print("Salario neto: $", self.calcular_salario_neto())


empleado1 = Empleado(48, 5000, 12.5)
empleado1.mostrar_resultados()
````

# **Ejercicio Propuesto N° 14**

Elabore un algoritmo que lea un número y obtenga su cuadrado y su cubo.


````python
class Numero:
    def __init__(self, valor):
        self.valor = valor

    def calcular_cuadrado(self):
        return self.valor ** 2

    def calcular_cubo(self):
        return self.valor ** 3

    def mostrar_resultados(self):
        print("Número:", self.valor)
        print("Cuadrado:", self.calcular_cuadrado())
        print("Cubo:", self.calcular_cubo())


num = float(input("Ingrese un número: "))

objeto_numero = Numero(num)
objeto_numero.mostrar_resultados()
````

# **Ejercicio Propuesto N° 17**

Dado el radio de un círculo. Haga un algoritmo que obtenga el área del círculo y la longitud
de la circunferencia.

````python
import math

class Circulo:
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * self.radio ** 2

    def longitud(self):
        return 2 * math.pi * self.radio

    def mostrar_resultados(self):
        print("Radio:", round(self.radio , 3))
        print("Área del círculo:", round(self.area(),3))
        print("Longitud de la circunferencia:", round(self.longitud(),3))


radio = float(input("Ingrese el radio del círculo: "))

mi_circulo = Circulo(radio)
mi_circulo.mostrar_resultados()
````
